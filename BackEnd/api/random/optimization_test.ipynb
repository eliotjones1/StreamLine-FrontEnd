```python
movies = list(providers.keys())
services = list(some_services)
costs = prices
# providers
# budget
```

```python
total = 0
for key in prices:
    total += prices[key]
print(total)
print(len(prices))

```

```python
x = {m: cp.Variable(boolean=True) for m in movies}
y = {s: cp.Variable(boolean=True) for s in services}

objective = cp.sum(1 - cp.vstack([x[m] for m in movies]))

services = list(set([s for v in providers.values() for s in v]))

constraints = [
    cp.sum(cp.vstack([y[s] * costs[s] for s in services])) <= budget,
]
for m in providers:
    constraints += [x[m] <= cp.sum(cp.vstack([y[s] for s in providers[m]]))]
    for s in services:
        if s in providers[m]:
            constraints += [y[s] <= cp.sum(cp.vstack([x[m] for m in providers]))]
    
problem = cp.Problem(cp.Minimize(objective), constraints)
problem.solve()

print("Movies included in the package:")
for m in movies:
    if x[m].value == 1:
        print("- ", m)

print("Streaming services included in the package:")
for s in services:
    if y[s].value == 1:
        print("- ", s)
```
